# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy.stats import multivariate_normal
import scipy
```

# Narysujmy gęstość dwuwymiarowego rozkładu Normalnego

```python
mu = np.zeros(2)
cov = np.eye(2)

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))
distr = multivariate_normal(mu, cov)
Z = distr.pdf(pos)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap = 'inferno')
fig.show()
```

# Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego

```python
plt.figure(figsize = (5,5))
plt.contourf(X, Y, Z, cmap='inferno')
plt.xlabel("x1")
plt.ylabel("x2");
```

## Narysujmy poziomice (contour plot) dla mieszanki gaussów

Rozpatrzmy mieszankę gaussów postaci
$$
\frac{1}{2}N(m_1,\Sigma_1)+\frac{1}{2}N(m_2,\Sigma_2).
$$
Proszę narysować poziomice dla przykładu gdy $m_1\neq m_2$, $\Sigma_1=\Sigma_2$

```python
mu1 = np.ones(2) * 3
mu2 = np.ones(2) * 0.5
cov1 = np.eye(2) * 5
cov2 = np.eye(2) * 2

tick = 7
x = np.linspace(-tick, tick, 100)
y = np.linspace(-tick, tick, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))
distr1 = multivariate_normal(mu1, cov1)
distr2 = multivariate_normal(mu2, cov2)
Z = 0.5 * distr1.pdf(pos) + 0.5 * distr2.pdf(pos)

plt.figure(figsize = (5,5))
plt.contourf(X, Y, Z, cmap='inferno')
plt.xlabel("x1")
plt.ylabel("x2");
```

# Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego

Proszę wylosować na dwa sposoby: korzystając z scipy.stats.multivariate_normal oraz numpy.random.multivariate_normal

```python
mn1 = scipy.stats.multivariate_normal(mean=np.zeros(2), cov=np.eye(2))
pts = mn1.rvs(size=5) # , random_state=12345
plt.plot(pts[:, 0], pts[:, 1], '.', markersize=20)
plt.axis('equal')
plt.grid()
plt.show()
pts
```

```python
pts2 = np.random.multivariate_normal(np.zeros(2), np.eye(2), size=5)
plt.plot(pts2[:, 0], pts2[:, 1], '.', markersize=20)
plt.axis('equal')
plt.grid()
plt.show()
pts2
```

# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy.stats import multivariate_normal
```

# Zad. 
Narysujmy gęstość dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
def draw_gaussian_distribution(n, cov):
    x = np.linspace(-3, 3, 100)
    y = np.linspace(-3, 3, 100)
    X, Y = np.meshgrid(x, y)
    pos = np.dstack((X, Y))
    mu = np.array(n)
    cov = np.array(cov)
    rv = multivariate_normal(mu, cov)
    Z = rv.pdf(pos)
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z)
    fig.show()



#https://stackoverflow.com/questions/38698277/plot-normal-distribution-in-3d
```

```python
draw_gaussian_distribution([0, 0], [[2,0],[0,1]])
```

```python
draw_gaussian_distribution([0, 0], [[1,0],[0,2]])
```
